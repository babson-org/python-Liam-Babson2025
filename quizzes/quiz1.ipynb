## Question 1: Diamond Printing

In the code cell below, modify the two `for` loop `range` statements to make the code print out a solid diamond.
Your output should look like:

```
  *
 ***
*****       # if height entered is 5
 ***
  *
```

````python
txt = "Please input a height, an odd integer between 3 and 15 inclusive: "
while True:
    try:
        height = int(input(txt))
    except ValueError:
        txt = 'Enter an integer: '        
    else:
        if(height < 3 or height > 15):
            txt = 'between 3 and 15 please: '
        elif height % 2 == 0:
            txt = 'an odd number please: '
        else:
            break

half = height // 2

# Top half of diamond
leading_spaces = half
for star_count in range(1, height + 1, 2):  # 1, 3, 5, 7, ... height
    print(" " * leading_spaces + "*" * star_count)
    leading_spaces -= 1

# Bottom half of diamond
leading_spaces = 1
for star_count in range(height - 2, 0, -2):  # height-2, height-4, ..., 1
    print(" " * leading_spaces + "*" * star_count)
    leading_spaces += 1

'''
## Explanation:

### Top Half - `range(1, height + 1, 2)`:
For height = 7:
- Generates: 1, 3, 5, 7
- Start at 1 (first odd)
- Stop at height + 1 (to include height)
- Step by 2 (odd numbers only)


**Visual:**

   *      (3 spaces, 1 star)
  ***     (2 spaces, 3 stars)
 *****    (1 space, 5 stars)
*******   (0 spaces, 7 stars)


### Bottom Half - `range(height - 2, 0, -2)`:
For `height = 7`:
- Start at `height - 2 = 5` (2 less than max)
- Stop at 0 (exclusive, so goes down to 1)
- Step by -2 (count down by 2)
- Generates: 5, 3, 1

**Visual:**
```
 *****    (1 space, 5 stars)
  ***     (2 spaces, 3 stars)
   *      (3 spaces, 1 star)
```

## Complete Output for height = 7:
```
   *
  ***
 *****
*******
 *****
  ***
   *
```

## Key Insights:

1. **Top half starts at 1, ends at height, increments by 2**
   - `range(1, height + 1, 2)`

2. **Bottom half starts at height - 2, ends at 1, decrements by 2**
   - `range(height - 2, 0, -2)`
   - We subtract 2 because we already printed `height` stars in the top half
   - We don't want to repeat the middle line

3. **Leading spaces**:
   - Top: starts at `half`, decreases
   - Bottom: starts at 1, increases

## Test with Different Heights:

**Height = 3:**
```
 *
***
 *
```

**Height = 5:**
```
  *
 ***
*****
 ***
  *
```

**Height = 9:**
```
    *
   ***
  *****
 *******
*********
 *******
  *****
   ***
    *

'''
````

## Question 2: Caesar Cipher Decryption

The program below will decrypt a message using the Caesar cipher. Julius needs to decode the message. Modify the code to try all possible shifts (1 to 25) to find the secret message.

Hint: Use `if 'python' in decode_msg:` to check for the correct decoding.

````python
import string

msg = 'xu ndj vti iwxh bthhpvt, ndj pgt bphitgxcv eniwdc'

alphabet = list(string.ascii_lowercase)

# TODO: create a loop from 1 to 25 to try all shifts
#       decode the message and print the one containing 'python'
#       The code below should be inside this loop

# CHANGE 1: Removed outer for i in range() loop - it was unnecessary
# CHANGE 2: Fixed range to go from 1 to 26 (exclusive), testing shifts 1-25
for shift in range(1, 26):  # TODO fix range
    
    shifted = [None] * 26
    for idx in range(26):
        new_idx = (idx - shift) % 26
        shifted[new_idx] = alphabet[idx]

    decode_msg = ''
    for character in msg:
        if character.isalpha():
            idx = shifted.index(character)
            decode_msg += alphabet[idx]
        else:
            decode_msg += character

    # TODO add code to check if python is in the decoded message
    # if it is print out the decoded message and the shift value
    
    # CHANGE 3: Added if statement to check if 'python' appears in decoded message
    if 'python' in decode_msg:
        # CHANGE 4: Added print statement showing shift value and decoded message
        print(f"Shift {shift}: {decode_msg}")
````
